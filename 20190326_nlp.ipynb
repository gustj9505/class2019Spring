### Text 열기

```python
raw = open(os.getcwd()+r'/06_01.txt', encoding = 'utf8').read()
#os에 있는 txt'까지의 path를 열고, utf8로 아마 인코딩이 되어 있을 것이다.
#인코딩 지정을 해주지 않으면 한글파일이 깨져서 나온다. 
raw
```

```python
print(type(raw))
print(len(raw))
print(raw[:75])
#불러온 raw의 타입, 길이, 75번째 글자까지를 출력해낸다. 
```

```python
tokens = nltk.word_tokenize(raw) #raw를 단어마다 쪼개서(tokenize) list로 저장한다. 
print(type(tokens))
print(len(tokens))
print(tokens[:10])
```

```python
text = nltk.Text(tokens) #tokens를 하나의 text로 묶는다. 
print(type(text))
print(text[:10]) #하나의 text로 묶었지만, list로 출력하는 것도 가능하다. string과 list의 평행이론
text.collocations() #text 안에서 이름 등과 같은 함께 자주 쓰이는 collocations를 출력하는 함수
```

```python
text.concordance('great', 79, 25)
#text 안에 있는 수많은 단어 중에서 great라는 단어가 79번째 자리에 오도록 25줄을 출력해낸다. 
#이 방법을 쓰면 또다른 crawling의 방법이 될 수 있다.
#162가지의 매치 중에서 25개만 출력을 한다. 
```

```python
text.concordance('my', 50, 5)
#790가지의 매치 중에서 5개만 출력한다. 
#my라는 단어가 50번째 자리에 오도록 5줄만 출력한다. 
```

### Frequent Distribution: 자주 나오는 횟수

```python
raw = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
raw
```

```python
fdist = nltk.FreqDist(raw)
print(fdist.most_common(5))
#fdist에서 가장 common한 alphabet 5가지를 뽑고, 그 횟수를 출력한다. pair로 출력된다.
```

```python
fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha())
#raw에 있는 character가 만약 alphabet이라면 소문자로 저장한 뒤, FreqDist함수에 넣어서 fdist에 저장한다. 
print(fdist.most_common(5))
```

```python
[char for (char, count) in fdist.most_common(5)]
#가장 흔하게 나오는 알파벳 다섯개를 출력한다. (char, count)의 pair에서 char만 출력한다.
```
