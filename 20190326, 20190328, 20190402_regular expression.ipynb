### treebank라는 corpus. 이것도 사전

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
wsj
#wsj = world street journal
```

```python
print(type(wsj))
#wsj의 타입도 list. 하나하나는 string이지만 전체는 string이 아니기 때문에 각 항목에 접근하려면 for문을 사용해야 한다. 
```

```python
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
print(wordlist[:10])
#[0-9]는 0부터 9까지의 숫자 중 아무거나 올 수 있음을 의미한다. 
#온점을 출력하고 싶다면 .가 아닌 \.로 입력해 주어야 한다. 그냥 .은 wildcard로, 아무런 character 하나를 출력한다. 
#+는 하나 이상, *는 0개 이상
```

```python
wordlist = [w for w in wsj if re.search('^[A-Z]+\$$', w)]
print(wordlist[:10])
#온점을 찍을 때와 마찬가지로, $를 직접 출력하고 싶다면 \$를 써줘야 한다. 그냥 $는 string의 끝을 의미한다.
```

```python
wordlist = [w for w in wsj if re.search('^[A-Z]+$', w)]
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[0-9]+$', w)]
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[A-Z0-9]+$', w)]
print(wordlist[:10])
wordlist = [w for w in wsj if re.search('^[A-Za-z]+$', w)]
print(wordlist[:10])
#[A-Z0-9]는 A-Z나 0-9 중 하나만 출력한다.
```

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)]
print(wordlist[:10])
#{4}는 정확히 4번 반복함을 의미한다.
```

```python
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
print(wordlist[:10])
#시스템 상에서 -에는 별다른 기능을 설정하지 않았으므로, 그냥 써줘도 -로 출력이 된다. 
#{3,5}는 3번에서 5번 반복함을 의미한다.
```

```python
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print(wordlist[:10])
#{5,}는 최소 5번 이상, {,6}는 최대 6번까지
#10개를 출력하려고 하더라도 wordlist에 5개밖에 없다면 5개만 출력된다.
```

```python
wordlist = [w for w in wsj if re.search('(ed|ing)$', w)]
print(wordlist[:10])
#ed나 ing로 끝나는 단어 10개
```

### FreqDist: Frequent Distribution: 자주 나오는 것. 

```python
re.findall('ab', 'ababdddabdsabfds')
#findall(a, b): b에서 a를 찾아서 다 뽑아내라. 
#4개이기 때문에 len()을 찾아보면 4
```

```python
[vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word)]
```

```python
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
fdist.most_common(10)
#r'는 raw text로, 코드를 작성한대로 다 출력해낸다. 각종 기능 없애는 방법. 
```

```python
word = 'supercalifragilisticexpialidocious'
```

```python
result = re.findall(r'[aeiou]', word)
print(result)
#word 안에서 모음을 찾아라. 
```

```python
result = re.findall('[aeiou](..)[aeiou]', word)
print(result)
#모음 사이에 character가 두 개 있는 경우를 찾아라. (..)부분이 출력될 것.
```

```python
result = re.findall('[aeiou].+[aeiou]', word)
print(result)
#모음과 모음 사이에 위치한 글자들이 출력될 것. 
result = re.findall('[^aeiou].+[^aeiou]', word) #greedy search
print(result)
#모음과 모음 사이에 글자 아무거나 오는 단어를 전부 찾기. 가능한 넓게 찾기
#.에 ()가 쳐져 있지 않기 때문에 앞 글자도 포함하게 된다. 이 경우에는 단어 자체가 된다.
#[^aeiou]에서 ^가 선택될 경우 aeiou중에서는 선택되지 않는다. 이 경우 .가 맨 첫 글자가 될 것. 
```

```python
result = re.findall('[^aeiou].?[^aeiou]', word)
print(result)
#?는 바로 앞 항목이 0개 아니면 1개. 있거나 없어도 된다. 가능한 좁게 찾기.
result = re.findall('[^aeiou].+?[^aeiou]', word) #reluctant search
print(result)
```

### 함수를 직접 만드는 방법

```python
regexp = '^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
#|를 기준으로 세 덩어리로 구분할 수 있다. 
# ^[AEIOUaeiou]+ 단어의 시작과 동시에 모음이 한 개 이상 나오는 경우
# [AEIOUaeiou]+$ 단어의 마지막 부분에 모음이 한 개 이상 나오는 경우
# [^AEIOUaeiou] 모음이거나 아닌 경우
def compress(word):
#compress라는 함수를 만들겠다. def 함수명(입력값), 입력값이 여러개라면 comma로 구분
#함수의 내용은 반드시 tab을 해주어야 한다. 아니면 에러가 발생할 수 있다. 
    pieces = re.findall(regexp, word)
    #string을 받아서 바로 regular expression을 적용시키는 함수이다. 
    return ''.join(pieces)
    #return: 반환값. 출력값. return값이 없다면 함수 내부적으로만 기능하고 출력을 하지 않는 것을 의미한다. 
    #pieces로 받은 것들을 join 하는데, ''빈 것을 이용하여 붙여라. 
[compress(w) for w in english_udhr[:75]]
#english_udhr에 있는 list를 차례로 하나씩 w로 받아서 compress 함수를 적용시킨다. 
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
print(rotokas_words)
```

```python
cvs = [cv for w in rotokas_words for cv in re.findall('[ptksvr][aeiou]', w)]
#rotokas_words의 단어들을 받아서 w로 하나씩 받는다. 그리고, 그 w 중에서 [ptksvr][aeiou]의 구조를 가진 것을 cv로 받는다. 
print(cvs)
```

```python
cfd = nltk.ConditionalFreqDist(cvs) #CFD words pairwise only
cfd.tabulate()
#tabulate(): 각 조건에 따라 표로 나타내주는 함수
```
